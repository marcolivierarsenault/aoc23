```python
from collections import Counter
```

```python
file1 = open('input.txt', 'r')
lines = file1.readlines()
```

```python
def get_hand_score(hand):
    if (len(hand) != 5):
        print("===ERROR not 5 card`s in hand===")
        return None

    dict_hand = dict(Counter(hand))
    if (len(dict_hand) == 1):
        return 0

    if (len(dict_hand) == 2):
        if sorted(list(dict_hand.values())) == [1, 4]:
            return 1
        if sorted(list(dict_hand.values())) == [2, 3]:
            return 2
    if (len(dict_hand) == 3):
        if sorted(list(dict_hand.values())) == [1, 1, 3]:
            return 3
        if sorted(list(dict_hand.values())) == [1, 2, 2]:
            return 4
    if (len(dict_hand) == 4):
            return 5
    return 6


assert get_hand_score(['A', 'A', 'A', 'A', 'A']) == 0
assert get_hand_score(['A', 'A', 'A', 'A', '1']) == 1
assert get_hand_score(['A', 'A', 'A', '2', '2']) == 2
assert get_hand_score(['A', 'A', 'A', '1', '2']) == 3
assert get_hand_score(['A', 'A', '1', '1', '2']) == 4
assert get_hand_score(['A', 'A', '4', '1', '2']) == 5
assert get_hand_score(['A', '4', '7', '1', '2']) == 6
```

```python
def p2_get_hand_score(hand):
    if (len(hand) != 5):
        print("===ERROR not 5 card`s in hand===")
        return None

    dict_hand = dict(Counter(hand))
    if ('J' not in dict_hand):
        if (len(dict_hand) == 1):
            return 0

        if (len(dict_hand) == 2):
            if sorted(list(dict_hand.values())) == [1, 4]:
                return 1
            if sorted(list(dict_hand.values())) == [2, 3]:
                return 2
        if (len(dict_hand) == 3):
            if sorted(list(dict_hand.values())) == [1, 1, 3]:
                return 3
            if sorted(list(dict_hand.values())) == [1, 2, 2]:
                return 4
        if (len(dict_hand) == 4):
                return 5
        return 6

    nb_of_J = dict_hand['J']
    if nb_of_J == 5:
        return 0
    del(dict_hand['J'])

    if (len(dict_hand) == 1):
            return 0

    if (len(dict_hand) == 2):
        if sorted(list(dict_hand.values())) == [1, 4-nb_of_J]:
            return 1
        if sorted(list(dict_hand.values())) == [2-nb_of_J, 3]:
            return 2
        if sorted(list(dict_hand.values())) == [2, 3-nb_of_J]:
            return 2
    if (len(dict_hand) == 3):
        if sorted(list(dict_hand.values())) == [1, 1, 3-nb_of_J]:
            return 3
    if (len(dict_hand) == 4):
            return 5


assert p2_get_hand_score(['A', 'A', 'A', 'A', 'A']) == 0
assert p2_get_hand_score(['A', 'A', 'A', 'A', '1']) == 1
assert p2_get_hand_score(['A', 'A', 'A', '2', '2']) == 2
assert p2_get_hand_score(['A', 'A', 'A', '1', '2']) == 3
assert p2_get_hand_score(['A', 'A', '1', '1', '2']) == 4
assert p2_get_hand_score(['A', 'A', '4', '1', '2']) == 5
assert p2_get_hand_score(['A', '4', '7', '1', '2']) == 6

assert p2_get_hand_score(['J', 'J', 'J', 'J', 'J']) == 0
assert p2_get_hand_score(['J', 'J', 'J', 'K', 'K']) == 0
assert p2_get_hand_score(['J', 'J', 'J', '4', 'K']) == 1
assert p2_get_hand_score(['J', 'K', '4', '4', 'K']) == 2
assert p2_get_hand_score(['J', 'J', '3', '4', 'K']) == 3
assert p2_get_hand_score(['J', '6', '3', '4', 'K']) == 5
```

```python
values = {
    'A': 14,
    'K': 13,
    'Q': 12,
    'J': 11,
    'T': 10,
    '9': 9,
    '8': 8,
    '7': 7,
    '6': 6,
    '5': 5,
    '4': 4,
    '3': 3,
    '2': 2,

}

part2_values = {
    'A': 14,
    'K': 13,
    'Q': 12,
    'T': 10,
    '9': 9,
    '8': 8,
    '7': 7,
    '6': 6,
    '5': 5,
    '4': 4,
    '3': 3,
    '2': 2,
    'J': 1,

}
```

```python
def card_value(cards):
    value = 0
    for i, cards in enumerate(cards):
        value += pow(10, 8 - (i*2)) * values[cards]
    return value

assert card_value(['A', '4', '7', 'K', '2']) == 1404071302

def p2_card_value(cards):
    value = 0
    for i, cards in enumerate(cards):
        value += pow(10, 8 - (i*2)) * part2_values[cards]
    return value

assert p2_card_value(['A', 'J', '7', 'K', '2']) == 1401071302

```

```python
cards = []
for line in lines:
    card_info = line.split()
    list_of_card = [*card_info[0]]
    cards.append(
        {
            'cards' : list_of_card,
            'bid': int(card_info[1]),
            'hand_score': get_hand_score(list_of_card),
            'cards_value': card_value(list_of_card)

        }
    )

counter = len(cards)
for i in range(7):
    cards_with_score = list(filter(lambda x: x['hand_score'] == i, cards))
    sorted_cards = sorted(cards_with_score, key=lambda x: x['cards_value'], reverse=True)
    for card in sorted_cards:
        card["final_position"] = counter
        card["final_score"] = counter * card["bid"]
        counter -= 1

print(f"Part 1: {sum(map(lambda x: x['final_score'], cards))}")
```

```python
cards2 = []
for line in lines:
    card_info = line.split()
    list_of_card = [*card_info[0]]
    cards2.append(
        {
            'cards' : list_of_card,
            'bid': int(card_info[1]),
            'hand_score': p2_get_hand_score(list_of_card),
            'cards_value': p2_card_value(list_of_card)

        }
    )

counter = len(cards2)
for i in range(7):
    cards_with_score = list(filter(lambda x: x['hand_score'] == i, cards2))
    sorted_cards = sorted(cards_with_score, key=lambda x: x['cards_value'], reverse=True)
    for card in sorted_cards:
        card["final_position"] = counter
        card["final_score"] = counter * card["bid"]
        counter -= 1

print(f"Part 2: {sum(map(lambda x: x['final_score'], cards2))}")
```
